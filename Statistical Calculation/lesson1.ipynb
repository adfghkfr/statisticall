```python
from numpy import linalg #numpy裡面的套件linalg，因為套件很多，必須要加入這行，使用時才不用加很長
import numpy as np 
import scipy.linalg as LA # scipy's linear algebra is better
 
X = np.array([[1,2,3],[4,5,6]]) # a 2x3 matrix
# check the information of X
Xdim = X.ndim
Xshape = X.shape
Xsize = X.size
print(Xdim)#2
print(Xshape)#(2,3)
print(Xsize)#6 #共6個元素
 
a = X[0, 1]
P = X * X # element by element multiplication
Q = X / X # element by element division
R = X @ X.T # XX'
print(a)
#2
print(P)
#[[ 1  4  9]
# [16 25 36]]
print(Q)
#[[1. 1. 1.]
# [1. 1. 1.]]
print(R)
#[[14 32]
# [32 77]]

#total_sum=np.sum(X) #從numpy取出
total_sum = X.sum() # X.mean()... #要X是numpy物件才可以取出
row_sum = X.sum(axis = 1) # sum over each row
col_sum = X.sum(axis = 0) # sum over each column
# col_sum = np.sum(X, axis = 0) 
print(total_sum)#21
print(row_sum)#[ 6 15]
print(col_sum)#[5 7 9]
 
A = np.array([[1, 1, 1],[2, 4, 6],[5, 7, 6]])
b = np.array([1, 2, 3])
x =  LA.inv(A) @ b  # Ax = b
xsol = LA.solve(A, b) # Ax = b
I = A @ LA.inv(A) # double check what LA.inv(A) is
print(A)
#[[1 1 1]
# [2 4 6]
# [5 7 6]]
print(b)
#[1 2 3]
print(x)
#[ 1.66666667 -1.33333333  0.66666667]
print(xsol)
#[ 1.66666667 -1.33333333  0.66666667]
print(I)
#[[ 1.00000000e+00  1.66533454e-16 -1.66533454e-16]
# [ 4.44089210e-16  1.00000000e+00 -6.66133815e-16]
# [ 2.22044605e-15  7.77156117e-16  1.00000000e+00]]

 
A_square = linalg.matrix_power(A, 2) # nth power of a matrix
A_square_root = LA.fractional_matrix_power(A, 0.5)
print(A_square)
#[[ 8 12 13]
# [40 60 62]
# [49 75 83]]
print(A_square_root)
#[[0.73878486-0.02707346j 0.24860816-0.02513898j 0.24396951+0.0236736j ]
# [0.42299787+0.73072319j 1.41039089+0.67851076j 1.53803548-0.63895957j]
# [1.29406597-0.66263317j 1.78200499-0.61528598j 1.91224586+0.57942024j]]
 
B = np.vstack((A, b)) # 4 x 3
C = np.hstack((A, b.reshape(3, 1))) # 3 x 4
print(B)
#[[1 1 1]
# [2 4 6]
# [5 7 6]
# [1 2 3]]
print(C)
#[[1 1 1 1]
# [2 4 6 2]
# [5 7 6 3]]
 
B_row = np.insert(A, 3, b, 0) # 3 is the position index; 0 is for row
B_col = np.insert(A, 3, b, 1) # 3 is the position index; 1 is for column
print(B_row)
#[[1 1 1]
# [2 4 6]
# [5 7 6]
# [1 2 3]]
print(B_col)
#[[1 1 1 1]
# [2 4 6 2]
# [5 7 6 3]]
```
